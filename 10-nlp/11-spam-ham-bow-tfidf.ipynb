```python
# Data cleaning and preprocessing
import re
import nltk 
```

```python
from nltk.corpus import stopwords 
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]

# input features
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
from nltk.stem import WordNetLemmatizer
# lammatised import 

lt=WordNetLemmatizer()
```

```python
corpus_lemmatised=[]

# input features
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    
    review=[lt.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus_lemmatised.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus_lemmatised
```

```python
tf=TfidfVectorizer(max_features=2500,ngram_range=(1,2)) 
```

```python
x=tf.fit_transform(corpus_lemmatised).toarray()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)   
```

```python
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred)) 
```

```python
print(accuracy_score(y_test,y_pred))
```
