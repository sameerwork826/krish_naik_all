```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<body>
    <div>
        <h1>Foo</h1>
        <p>Some intro text about Foo.</p>
        <div>
            <h2>Bar main section</h2>
            <p>Some intro text about Bar.</p>
            <h3>Bar subsection 1</h3>
            <p>Some text about the first subtopic of Bar.</p>
            <h3>Bar subsection 2</h3>
            <p>Some text about the second subtopic of Bar.</p>
        </div>
        <div>
            <h2>Baz</h2>
            <p>Some text about Baz</p>
        </div>
        <br>
        <p>Some concluding text about Foo</p>
    </div>
</body>
</html>
"""

header_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]
# header tag that we will use to split

html_splitter=HTMLHeaderTextSplitter(header_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
html_header_splits
```

```python
print(html_header_splits[0])
print(html_header_splits[1])
print(html_header_splits[2])     
print(html_header_splits[3])
print(html_header_splits[4])            
print(html_header_splits[5])
print(html_header_splits[6])   
# here we can see that the text is split into 3 parts based on the header tags
```

Vector Store DB

FAISS

Facebook ai similarity search faiis for efficient similartiy search and cllustering of dense vectors

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS 
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter 

loader=TextLoader("speech.txt")
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=200,chunk_overlap=30)
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings=OllamaEmbeddings(model="gemma:2b")

# here we made db of faiss

db=FAISS.from_documents(docs,embeddings)
db
```

```python
# do query for similarity search

query="What does the speaker beleive is the main reason the united states go into war"
result=db.similarity_search(query)
```

```python
result[0].page_content
```

As a retriever 

We can also convert the vector store into retriever clss this allows us to eaily use it in other LangChain methodds which largely work with retireivers

```python
retriever=db.as_retriever()
docs=retriever.invoke(query)
```

```python
docs[0].page_content
# here we used the vector store to basically
# do simi search without doing it 
```

```python
# now with similarity search method 

docs_and_sscore=db.similarity_search_with_score(query)

docs_and_sscore
# the score is attached below nearest will be first 
# according to query
```

```python
# can we just pass vecotr instead of sentence

embedding_vector=embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
# how to store vector db into local

db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index",embeddings,allow_dangerous_deserialization=True)
new_db
```

```python
docs=new_db.similarity_search(query)
```

```python
docs[0].page_content
```

### Chroma DB

Chroma is ai native open sourcfe vecotr database focused on developer productivity 

```python
from langchain_chroma import Chroma 
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader=TextLoader("speech.txt")
data=loader.load()
```

```python
data[0].page_content
```

```python
# using spliiter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
splits=text_splitter.split_documents(data)
```

```python
embedding=OllamaEmbeddings(model="gemma:2b")
vectordb=Chroma.from_documents(documents=splits,embedding=embedding,persist_directory="./chroma_db")
```

```python
db2=Chroma(persist_directory="./chroma_db",embedding_function=embedding)
docs=db2.similarity_search(query)
print(docs[0].page_content)
```

```python
# retriever option

retreiver=vectordb.as_retriever()
retriever.invoke(query)[0].page_content
```
